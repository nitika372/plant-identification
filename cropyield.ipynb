```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
crop=pd.read_csv("crop_yield.csv")
crop.head(4)
```

```python
import joblib
```

```python
crop.loc[:,'Crop']= crop.loc[:,'Crop'].str.strip()
```

```python
crop_enc= LabelEncoder()
crop_enc.fit(crop.loc[:,'Crop'])
joblib.dump(crop_enc, 'crop_enc.pkl')
crop.loc[:,'Crop']=crop_enc.transform(crop.loc[:,'Crop'])
```

```python
crop.loc[:,'Crop']
```

```python
crop_enc.classes_
```

```python
crop_enc.transform(['Coconut'])[0]
```

```python
crop.loc[:,'Season']= crop.loc[:,'Season'].str.strip()
```

```python
season_enc= LabelEncoder()
season_enc.fit(crop.loc[:,'Season'])
joblib.dump(season_enc, 'season_enc.pkl')

crop.loc[:,'Season']= season_enc.transform(crop.loc[:,'Season'])
```

```python
season_enc.classes_
```

```python
state_enc= LabelEncoder()
state_enc.fit(crop.loc[:,'State'])
joblib.dump(state_enc, 'state_enc.pkl')

crop.loc[:,'State']= state_enc.transform(crop.loc[:,'State'])
```

```python
crop
```

```python
crop
```

```python
# Aplly the random forest regression............
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
x=crop.iloc[:,:-1]
y=crop.iloc[:,-1]
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=20)
```

```python
rf=RandomForestRegressor(n_estimators=100)
rf.fit(x_train,y_train)
```

```python
y_predict=rf.predict(x_test)
```

```python
# using the metrics..............
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
predict=r2_score(y_test,y_predict)
predict
```
